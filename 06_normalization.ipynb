```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data.drop('class',axis=1),data['class'],test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
```

```python
x_train_normalized=scaler.transform(x_train)
x_train_normalized=pd.DataFrame(x_train_normalized)
x_test_normalized=scaler.transform(x_test)
```

```python
fig,(sub_fig1,sub_fig2)=plt.subplots(nrows=1,ncols=2,figsize=(10,4))
sub_fig1.set_title('Before Scaling')
sub_fig1.scatter(x_train['Alcohol'],x_train['Malic acid'],c=y_train)
sub_fig2.set_title('After Scaling')
sub_fig2.scatter(x_train_normalized[0],x_train_normalized[1],c=y_train)
plt.show()
```

```python
fig,(sub_fig1,sub_fig2)=plt.subplots(nrows=1,ncols=2,figsize=(10,4))
sub_fig1.set_title('Before Scaling')
sns.kdeplot(x_train['Alcohol'],)
sub_fig2.set_title('After Scaling')
sns.kdeplot(x_train_normalized[0])
plt.show()
```
