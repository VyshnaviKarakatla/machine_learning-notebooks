```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
base_tree=DecisionTreeClassifier(max_depth=1)
```

```python
data=load_iris()
```

```python
x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=0.3,random_state=40)
```

```python
ada=AdaBoostClassifier(estimator=base_tree,n_estimatros=50,random_state=40)
```

```python
ada.fit(x_train,y_train)
```
