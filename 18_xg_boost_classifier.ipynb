```python
from xgboost import XGBClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data=load_iris()
```

```python
x_train,x_test,y_train,y_test=train_test_split(data.data,data.target,test_size=0.3,random_state=40)
```

```python
xgb=XGBClassifier(n_estimators=50,learning_rate=0.1,max_depth=3,random_state=40,user_label_encoder=False,eval_metric='mglogloss')
```

```python
xgb.fit(x_train,)
```
