```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({

    'height':[150,160,170,100,190],
    'weight':[50,60,70,80,90]
})
df
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaled_data=scaler.fit_transform(df)
```

```python
scaled_df=pd.DataFrame(scaled_data)
scaled_df
```
